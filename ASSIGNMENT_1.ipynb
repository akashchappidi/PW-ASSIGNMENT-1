5Q) What is mutability in python ?
- Mutability in Python refers to the ability of an object to change its value after it's created. Mutable objects can be modified directly, while immutable objects cannot.

6Q) Why are lists mutable, but tuples are immutable?
- Lists are mutable, allowing changes to their elements after creation. Tuples are immutable, preserving their contents once defined, ensuring data consistency.

7Q) What is the difference between "==" and "is" operators in python ?

 - The "==" operator checks for value equality, comparing the contents of two objects. The "is" operator checks for object identity, determining if two variables refer to the same object in memory.

 8Q) What is type casting in python ?
  - Type casting, or type conversion, is the process of converting one data type into another. Python supports both implicit (automatic) and explicit (manual) type casting. Implicit casting occurs when Python automatically converts data types to prevent data loss, while explicit casting requires using built-in functions like int(), float(), and str() to specify the desired conversion.

  9Q) What are logical operators in python ?
  - Logical operators in Python are and, or, and not. They combine or negate boolean expressions, returning True or False based on the given conditions.

  10Q) What is difference between implicit and explicit type casting?
  - Implicit Type Casting:
  Python automatically converts data types to prevent data loss.
  For example, when an integer is added to a float, Python implicitly converts the integer to a float.

   - Explicit Type Casting:
The programmer manually converts a data type using functions like int(), float(), and str(). This is useful for specific conversions or when automatic conversion might lead to unexpected results.

11Q) What is the purpose of conditional statements in python?
- Conditional statements in Python allow your program to make decisions based on specific conditions. They control the flow of execution, executing different code blocks depending on whether a condition is true or false. This enables your program to respond dynamically to various inputs and situations.

12Q) How does the elif statement work?
- The elif statement checks multiple conditions sequentially after an if statement. If the if condition is false, it moves to the first elif and continues checking until a true condition is found or all elif conditions are false.

13Q) What is difference between for and while loops ?
- For Loop:

Iterates over a sequence (like a list, tuple, or string) a specific number of times.
The loop terminates when the sequence is exhausted.

 - While Loop:

Executes a block of code repeatedly as long as a given condition is true.
The loop continues until the condition becomes false.
